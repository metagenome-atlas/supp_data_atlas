```python
def Label(Letter,**kwargs):
    
    default_args=dict(x=0,y=10,size=12, weight='bold')
    default_args.update(kwargs)
    return Text(Letter,**default_args )
                   

scale_factor=0.3



from svgutils.compose import *


Figure_name='Figure2.svg'

Figure("170mm", "14cm", 
        Panel(
              SVG('Figures/completness_contamination.svg').scale(0.7),
              Label("A")
             ),
        Panel(
              SVG('Figures/Mapping_ratio.svg' ).scale(0.7),
              Label("B")
             ).move(240, 0),
        Panel(
              SVG('Figures/tree.svg').scale(0.5).move(10,10),
              Label("C"),
              Text("Cecum",125,250,size=10 ),
              Text("Feces",125,280,size=10 )
            
             ).move(355, 0),
       
        Panel(
              SVG('Figures/heatmap.svg').scale(0.7),
              Label("D",y=0),
              Text("Cecum",125,75,size=10 ),
              Text("Feces",235,75,size=10 )
             ).move(0, 190),
        Panel(
              SVG('Figures/PCA_functions.svg').scale(0.7),
              Label("E",y=0)
             ).move(310, 285),
       
       #Grid(50,50)
        ).save(Figure_name),
        
    
    
    

embedSVG(filename=Figure_name)
```
